```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import dates as mpl_dates
%matplotlib inline
df = pd.read_csv('Energy_Dataset.csv')
df.Timestamp = pd.to_datetime(df.Timestamp)
rel_df = df.drop(labels=['Postcode','Customer', 'Price_AUD/MWh', 'Total_Demand_MWh', 'Latitude','Longitude' ], axis = 1)
cust = rel_df.pivot_table(index='Timestamp',aggfunc='mean')
```

```python
i = 119 # Custumer id

winter =['Winter','2012-07-15']
spring =['Spring','2012-10-15']
summer = ['Summer','2013-01-15']
autumn = ['Autumn','2013-04-15'] 



```

```python
# winter
sol = cust.Solar_Production_kWh.loc[winter[1]]
cons = cust.Energy_Consumption_kWh.loc[winter[1]]
heat = cust.Heating_Consumption_kWh.loc[winter[1]]

plt.plot(sol[sol>0], marker='o', ls='', label='Solar Energy Production')
plt.plot(cons[cons>0], marker='^', ls='',label='Energy Consumption' )
plt.plot(heat[heat>0], marker='D', ls='', label='Heating Load')

plt.ylabel('Energy(kWh)',fontsize=12)
plt.xlabel(spring[1],fontsize=12)
plt.gca().set_ylim([0,1.5])

plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%H:%M')
plt.gca().xaxis.set_major_formatter(date_format)

plt.legend(loc=1, fontsize =12, frameon=False, ncol=1)
plt.title(winter)
plt.tight_layout()
plt.show()
```
